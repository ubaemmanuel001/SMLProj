```python
# fill MonthlyIncome with the median
median_monthly_income = df['MonthlyIncome'].median()
# median_monthly_income
df['MonthlyIncome'].fillna(median_monthly_income, inplace=True)
```

```python
# Impute missing values for NumberOfDependents with the mode
mode_dependents = df['NumberOfDependents'].median()
# mode_dependents
df['NumberOfDependents'].fillna(mode_dependents, inplace=True)
```
